## Convert JSON to CSV

```python
import pandas as pd

# Read JSON files into DataFrames
df_fashion = pd.read_json('CS-THESIS/data/raw_data/AMAZON_FASHION_5.json')
df_phones = pd.read_json('CS-THESIS/data/raw_data/Cell_Phones_and_Accessories_5.json')
df_grocery = pd.read_json('CS-THESIS/data/raw_data/Grocery_and_Gourmet_Food_5.json')

df_fashion.to_csv('fashion.csv', index=False)
df_phones.to_csv('phones.csv', index=False)
df_grocery.to_csv('grocery.csv', index=False)
```
